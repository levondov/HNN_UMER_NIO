# Building Models of UMER

The goal here is to build an accurate model of UMER based on single particle tracking code AT, pic code WARP, and experimental measurements from UMER.

### FODO ring - simplest model

A simple model of UMER can be built using an 18 fold symmetry lattice. Each section/period is 64 cm long with 4 quads placed 16 cm apart in a DOFODOF configuration.

![umer_fodo_1cell](figures/fodo_1cell.svg)

The dot/circle represents where a typical UMER BPM would go. Putting together 18 periods we get a ring that has a 11.52 m circumference. 

![umer_fodo_1cell](figures/fodo_ring.svg)

In order to match certain accelerator parameters we need to adjust our quadrupole's effective length and focusing strength. These variables are important if one is to keep a beam matched and measure the correct tune. Typical machines employ a hardedge model using the middle of the fringe fields to create an effective length of a magnet. This works well if the length of the magnet is long compared to the fringe fields. The problem for UMER is that this condition is not true. Our field profiles are shaped more like gaussians than flat tops, making it difficult to easily find an effective length.

```python
# load maglie data
dt = np.load('magnets/RQUAD3.MAGLI.pkl')
#dt = np.load('magnets/RQUAD3S.MAGLI.pkl')
#dt = np.load('magnets/RQUAD3G.MAGLI.pkl')

bx = np.array([])
by = np.array([])
zpts = dt['magbx'].shape[-1]
for i in range(zpts):
    bx = np.append(bx,-1*dt['magbx'][25][12][i])
    by = np.append(by,-1*dt['magby'][12][25][i])    
```

```python
sgL = lambda x: 1/(1+np.exp(-10*(1+x)))
sgR = lambda x: 1/(1+np.exp(-10*(1-x)))

plt.figure(figsize=(16,3))

plt.subplot(1,2,1)
# plot field profile
x = np.linspace(-2,0,100)
plt.plot(x,sgL(x),color='k')
x = np.linspace(0,2,100)
plt.plot(x,sgR(x),color='k',label='field profile')
# plot hard edge approx
plt.plot([-1,-1,1,1],[0,1,1,0],color='red',linestyle='--',label='hard edge')
plt.title('Hard edge approximations - typical magnet')
plt.xlabel('Z (through magnet)')
plt.ylabel('$B_y$ (a.u.)')
plt.xticks([])
plt.yticks([])
plt.legend();

plt.subplot(1,2,2)
# plot field profile
x = np.linspace(-1,1,zpts)
plt.plot(x,bx,color='k')
# plot hard edge approx
plt.plot([-0.1,-0.1,0.1,0.1],[0.0,1.2*np.max(bx),1.2*np.max(bx),0.0],color='blue',linestyle='--', label='option 1')
plt.plot([-0.2,-0.2,0.2,0.2],[0.0,np.max(bx),np.max(bx),0.0],color='red',linestyle='--', label='option 2')
plt.plot([-0.5,-0.5,0.5,0.5],[0.0,0.5*np.max(bx),0.5*np.max(bx),0.0],color='green',linestyle='--', label='option 3')
plt.title('Hard edge approximations - UMER magnet')
plt.xlabel('Z (through magnet)')
plt.ylabel('$B_y$ (a.u.)')
plt.xticks([])
plt.yticks([])
plt.legend();
```

Historically, options 2 and 3 have been used at UMER to create models of the machine. The table below shows typical values for both the magnet strengths and effective lengths used. Note values are normalized to physical lengths and measured peak fields.

```python
maglst = np.array([['Ring Quad',4.65,0.962,3.609,0.8354,0.7208],['QR1',5.40,1.111,1.010,0.8965,0.7224],
                  ['YQ',5.40,1.111,1.110,0.8965,0.7224],['Dipole',4.437,0.8607,5.216,-1,-1],['PD',4.40,1.177,0.382,-1,-1]])

df = pd.DataFrame(maglst,columns=['Magnet','Length (cm)','Effective Length','PS (G/cm/A)','Effective PS 1','Effective PS 2'])
df=df.reindex(columns=['Magnet','Length (cm)','PS (G/cm/A)','Effective Length','Effective PS 1','Effective PS 2'])
df.head()
```

As an example, we use settings from row 0 and run a beam through a single cell lattice (from above). This is done both in a PIC code (WARP) and a single particle code (Accelerator Toolbox, AT). We inject the particle with a 1 mm offset in both planes.

```python
df = pd.read_csv('data/warp/1cell_full.txt')
dfs = pd.read_csv('data/matlab/1cell.txt',names=['zbeam','xcent','ycent'])

plt.figure(figsize=(20,4))
ax1=plt.subplot(1,2,1)
df.plot(x='zbeam',y=['xcent','ycent'],ax=ax1,label=['x WARP','y WARP'])
dfs.plot(x='zbeam',y=['xcent','ycent'],ax=ax1,marker='.',color='k',linestyle='',label=['x AT','y AT'])
plt.ylabel('Centroid Position')
plt.title('Single cell positions')
ax2=plt.subplot(1,2,2)
df.plot(x='zbeam',y=['2*xrms','2*yrms'],ax=ax2)
plt.ylabel('Envelope')
plt.title('Single cell envelopes');

df = pd.read_csv('data/warp/18cell_full.txt')
dfs = pd.read_csv('data/matlab/18cell.txt',names=['zbeam','xcent','ycent'])

plt.figure(figsize=(20,4))
ax1=plt.subplot(1,2,1)
df.plot(x='zbeam',y=['xcent','ycent'],ax=ax1,label=['x WARP','y WARP'])
dfs.plot(x='zbeam',y=['xcent','ycent'],ax=ax1,marker='.',color='k',linestyle='',label=['x AT','y AT'])
plt.ylabel('Centroid Position')
plt.title('1 turn positions')
ax2=plt.subplot(1,2,2)
df.plot(x='zbeam',y=['2*xrms','2*yrms'],ax=ax2)
plt.ylabel('Envelope')
plt.title('1 turn envelopes');
```

We can directly calculate the different frequencies both for the single cell and 1 turn lattices.

```python
from naff import *
v         = 5.845514528536631e07
revf      = 1/(0.64/v); # 64cm cell * 18 cells
tt        = 5e-3/v; # time between each point (each quad)

freqs,amps = naff(df[['xcent','ycent']].values)
freqs_centroid = freqs/tt/revf

freqs,amps = naff(df[['2*xrms','2*yrms']].values)
freqs_env = freqs/tt/revf
```

```python
freqs_lst = np.array([[0.962, 0.8354, freqs_centroid[0], freqs_centroid[1], freqs_centroid[0]*18, freqs_centroid[1]*18,
                      freqs_env[0], freqs_env[1], freqs_env[0]*18, freqs_env[1]*18],
                      [1.1105, 0.7208, 6.1669/18, 6.1675/18, 6.1669, 6.1675, -1, -1, -1, -1]])
names = ['Effective Length', 'Effective PS', '$Q_x$ 1 cell','$Q_y$ 1 cell','$Q_x$ 1 turn','$Q_y$ 1 turn',
         '$\sigma_x$ 1 cell','$\sigma_y$ 1 cell','$\sigma_x$ 1 turn','$\sigma_y$ 1 turn']
df_freq = pd.DataFrame(freqs_lst, columns=names)
df_freq.head()
```

Scanning a range of effective lengths and peak strengths will give us a picture of what values to use in order to get a ball park tune in the range of 6-7. The previous two points are also plotting as black dots. The scan is also done using the single particle model.

### Adding in Dipoles

Next we add in rectangular dipoles to our lattice and investigate the edge focusing effects on our tune. No earth field is included in our models.

![umer_fodo_1cell](figures/fodo_dipo_1cell.svg)

```python
df = pd.read_csv('data/warp/1cell_dipo_full.txt')
df1 = pd.read_csv('data/warp/1cell_bgrd_full.txt')
dfs = pd.read_csv('data/matlab/1cell_dipo.txt',names=['zbeam','xcent','ycent'])

plt.figure(figsize=(20,4))
ax1=plt.subplot(1,2,1)
df.plot(x='zbeam',y=['xcent','ycent'],ax=ax1,label=['x WARP','y WARP'])
df1.plot(x='zbeam',y=['xcent','ycent'],ax=ax1,label=['x WARP bgrd','y WARP bgrd'],linestyle='--')
dfs.plot(x='zbeam',y=['xcent','ycent'],ax=ax1,marker='.',color='k',linestyle='',label=['x AT','y AT'])
plt.ylabel('Centroid Position')
plt.title('Single cell positions')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
#ax2=plt.subplot(1,2,2)
#df.plot(x='zbeam',y=['2*xrms','2*yrms'],ax=ax2)
#plt.ylabel('Envelope')
#plt.title('Single cell envelopes');
```

Here we start to see a divergence in the models. Results are summarized below. We use quad values from row 0 of the previous table.

```python
freqs_lst = np.array([['AT',0.8606, 0.9969, 6.3208, 6.3208, 6.3161, 7.0211],['AT earth',0.8606,0.9969, 6.3208, 6.3208, 6.3123, 6.8265],
                      ['WARP dipo', 0.8606, 0.9969, 6.3277, 6.3274, 6.9323, 6.3276], ['WARP dipo earth', 0.8606, 0.9969, 6.3277, 6.3274, 6.8206, 6.3247], 
                      ['WARP bgrd', 0.8606, 0.9969, 6.3277, 6.3274, 6.4285, 6.7628], ['WARP bgrd earth', 0.8606, 0.9969, 6.3277, 6.3274, 6.4105, 6.6044]])

names = ['Model', 'Eff. Length', 'Eff. PS', '$Q_x$','$Q_y$', '$Q_x$ w/ dipo','$Q_y$ w/ dipo']
df_freq = pd.DataFrame(freqs_lst, columns=names)
pd.set_option('display.max_colwidth', 100)
df_freq
```
